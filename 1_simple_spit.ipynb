```python
file_path = 'constitutionrf.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()


def simple_text_splitter(text: str, chunk_length: int = 80):
    chunks = [text[i:i + chunk_length] for i in range(0, len(text), chunk_length)]
    return chunks


chunks = simple_text_splitter(text)

print(f"Всего чанков:", len(chunks))
print("Первые N чанков:", chunks[:10])
```

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator="\n", chunk_size=80, chunk_overlap=10)

chunks = text_splitter.create_documents([text])

print(f"Всего чанков:", len(chunks))
print("Первые N чанков:", chunks[:10])
```
